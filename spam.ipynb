```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, confusion_matrix
data={
    'message': [
    'you won lottery',
    'extra income',
    'won cash prize',
    'limited time offer',
    'get rich quick',
    'buy 1 get 1 free',
    'you won 1 crore coupoun',
    'get loan of 100 crore today',
    'congratulations you won lottery',
    'do not miss this',
    'join this zoom meeting',
    'reminder',
    'upgrade your plan',
    'your otp is',
    'update your payment method',
    'Bank statement',
    'verify its you',
    'invoice',
    'payment confirmed'
    ],
    'label': [
    'spam',
    'spam',
    'spam',
    'spam',
    'spam',
    'spam',
    'spam',
    'spam',
    'spam',
    'spam',
    'not spam',
    'not spam',
    'not spam',
    'not spam',
    'not spam',
    'not spam',
    'not spam',
    'not spam',
    'not spam'
    ]
}
df=pd.DataFrame(data)
df['label']=df['label'].map({'spam':1,'not spam':0})
vectorizer = CountVectorizer()
x=vectorizer.fit_transform(df['message'])
y=df['label']
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
madel = MultinomialNB()
madel.fit(x_train,y_train)
y_pred=madel.predict(x_test)
accuracy=accuracy_score(y_test,y_pred)
confusion=confusion_matrix(y_test,y_pred)
print(f'accuracy:{accuracy}')
new_email = ['hlo']
new_email_vectorized = vectorizer.transform(new_email)
prediction = madel.predict(new_email_vectorized)
if prediction[0] == 1:
    print('spam')
else:
    print('not spam')
```
